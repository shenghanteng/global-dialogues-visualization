```python
def f_coord(df, i=0, iframe=0):
    S = {'Negative': -1, 'Neutral':0, 'Positive': 1, '--': -0.01}
    if iframe==0:
        x = S[df.iloc[i,:][28]]
        y = (S[df.iloc[i,:][20]]*1+S[df.iloc[i,:][60]]*5)/6.
        z = (S[df.iloc[i,:][40]]*1+S[df.iloc[i,:][48]]*3+S[df.iloc[i,:][82]]*6)/10.
    elif iframe==1:
        x = S[df.iloc[i,:][15]]
        y = (S[df.iloc[i,:][24]]*2+S[df.iloc[i,:][64]]*6)/8.
        z = S[df.iloc[i,:][44]]
    elif iframe==2:
        x = S[df.iloc[i,:][72]]
        y = (S[df.iloc[i,:][32]]*3+S[df.iloc[i,:][36]]*4+
             S[df.iloc[i,:][68]]*7+S[df.iloc[i,:][76]]*8)/22.
        z = (S[df.iloc[i,:][52]]*4+S[df.iloc[i,:][56]]*5)/9.
    else:
        x = (S[df.iloc[i,:][15]]*1+S[df.iloc[i,:][28]]*2+S[df.iloc[i,:][72]]*3)/6.
        y = (S[df.iloc[i,:][20]]*1+S[df.iloc[i,:][24]]*2+S[df.iloc[i,:][32]]*3+
             S[df.iloc[i,:][36]]*4+S[df.iloc[i,:][60]]*5+S[df.iloc[i,:][64]]*6)/21.
        z = (S[df.iloc[i,:][40]]*1+S[df.iloc[i,:][44]]*2+S[df.iloc[i,:][48]]*3+
             S[df.iloc[i,:][52]]*4+S[df.iloc[i,:][56]]*5+S[df.iloc[i,:][82]]*6)/21.
    return x, y, z

def f_prop(df, i=0):
    S_age = {'Less than 18': 0, '18-25': 1, '26-35': 2, '36-45': 3, '46-55': 4, 
         '56-65': 5, '65+': 6}
    S_gender = {'Female': 0, 'Male': 1, 'Non-binary': 2, 'Other / prefer not to say': 3}
    S_live = {'Suburban': 0, 'Urban': 1, 'Rural': 2}
    S_rel = {'Buddhism': 0, 'Christianity': 1, 'Hinduism': 2, 'Islam': 3, 'Judaism': 4,
             'Sikhism': 5, 'I do not identify with any religious group or faith': 6,
             'Other religious group': 7}
    return S_age[df.iloc[i,3]], S_gender[df.iloc[i,4]], S_live[df.iloc[i,5]], S_rel[df.iloc[i,7]]

def write_dump(df, fname='dump'):
    Frame = {}
    for fr in range(4):
        if fr not in Frame:
            Frame[fr]=[]
        for i in range(len(df)):
            Frame[fr].append(f'{i} 0 '+'{:f} {:f} {:f} '.format(*f_coord(df, i, fr))+'{:d} {:d} {:d} {:d} '.format(*f_prop(df, i)))
    
    with open(fname, 'w') as fw:
        for fr in range(4):
            fw.write('ITEM: TIMESTEP\n')
            fw.write(f'{fr}\n')
            fw.write('ITEM: NUMBER OF ATOMS\n')
            fw.write(f'{len(Frame[fr])}\n')
            fw.write('ITEM: BOX BOUNDS pp pp pp\n')
            fw.write('-1 1\n')
            fw.write('-1 1\n')
            fw.write('-1 1\n')
            fw.write('ITEM: ATOMS id type xu yu zu age gender live rel\n')
            for d in Frame[fr]:
                fw.write(d+'\n')
```

```python
write_dump(df, fname='dump_gd1')
```
